#### Логарифмический масштаб

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
masses = [0.055274, 0.815, 1.0, 0.107, 317.8, 95.0, 14.6, 17.147]
```

```python
# Линейный масштаб
plt.plot(planets, masses)
plt.show()
```

```python
masses[planets.index('Earth')] / masses[planets.index('Mercury')]
```

```python
# Логарифмический масштаб на вертикальной оси
plt.plot(planets, masses)
plt.yscale(value='log')
plt.show()
```

```python
planet_info = pd.Series(masses, index=planets)
planet_info
```

```python
planet_info.plot()
plt.show()
```

```python
# Столбчатая диаграмма
planet_info.plot('bar')
plt.show()
```

```python
planet_info.plot(kind='bar', logy=True)
plt.show()
```

```python
planet_info.plot(kind='barh', logx=True)
plt.show()
```

### FIGURE И AXES

Figure и axes - основные объекты библиотеки Matplotlib
объекты axes располагаются внутри figure

```python
x = np.linspace(-5, 5, 101)
x2 = x**2
x3 = x**3
```

### Изпользуется структурный подход

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2)
plt.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, x2)
```

### Объекты figure и axes

```python
# Создание объектов figure и axes
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax
ax1.plot(x, x2)
ax2.plot(x, x3)
```

#### Расположение axes внутри объекта figure

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
```

#### Свойства объекта figure

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8.25, 3.5)
plt.subplots_adjust(wspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
print(ax.shape)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 6)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 6)
plt.subplots_adjust(hspace=0.25)
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
ax3.plot(x, y)
ax4.plot(x, z)
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

### Редактирование объекта axes

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График sin(x)')

ax4.plot(x, z)
ax4.set_title('График cos(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, sharey=True)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График sin(x)')

ax4.plot(x, z)
ax4.set_title('График cos(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')
ax1.set_xlim([-3, 3])

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График sin(x)')

ax4.plot(x, z)
ax4.set_title('График cos(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
